```python
from aocd.models import Puzzle
from typing import Tuple, Dict
from collections import defaultdict, deque
from functools import lru_cache
import re

puzzle = Puzzle(year=2020,day=7)
# print(puzzle.input_data)
```

```python
lines = puzzle.input_data.splitlines()

bag_dict = defaultdict(set)

for line in lines:
    parent, children = line.split('bags contain')
    for bagstr in children.split(','):
        if 'no other bags' in bagstr:
            bag_dict[parent.strip()] = []
        else:
            num_bags, color = bagstr.replace('bags', '').replace('bag', '').rstrip('. ').split(maxsplit=1)
            bag_dict[parent.strip()].add((int(num_bags), color))

# a 
# @lru_cache
def contains_color(color: str, target_color: str) -> bool:
    for _, child_color in bag_dict[color]:
        if child_color == target_color:
            return True
        if contains_color(child_color, target_color):
            return True
    return False

@lru_cache
def number_of_bags(color: str) -> int:
    if len(bag_dict[color]) == 0:
        return 0    
    total = sum([x*(1+number_of_bags(y)) for x,y in bag_dict[color]])
    return total

answer_a = sum([contains_color(x,'shiny gold') for x in list(bag_dict.keys())]) 
answer_b = number_of_bags('shiny gold')
```

```python
puzzle.answer_a = answer_a
puzzle.answer_b = answer_b
```
