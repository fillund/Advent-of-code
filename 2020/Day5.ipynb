```python
from aocd.models import Puzzle
from typing import Tuple, Dict
from collections import defaultdict, deque
from functools import lru_cache
import re

puzzle = Puzzle(year=2020,day=5)
print(puzzle.input_data)
```

```python
replacements = str.maketrans({'F':'0', 'B':'1', 'L':'0', 'R':'1'})
binary_lines = [x.translate(replacements) for x in puzzle.input_data.splitlines()]

def parse_row_col(line:str) -> Tuple[int, int]:
    row = int('0b'+line[:7], base=2)
    col = int('0b'+line[7:], base=2)
    return row,col

def calc_seatID(row, col) -> int:
    return row * 8 + col
seatIDs = [calc_seatID(*parse_row_col(x)) for x in binary_lines]
sortedIDs = sorted(seatIDs)

lastID = sortedIDs[0]
for ID in sortedIDs[1:]:
    if ID-lastID != 1:
        break
    lastID = ID
answer_a = max(seatIDs)
answer_b = ID-1
```

```python
puzzle.answer_a = answer_a
puzzle.answer_b = answer_b
```
