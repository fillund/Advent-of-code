```python
from aocd.models import Puzzle
from typing import Tuple, Dict
from collections import defaultdict, deque
from functools import lru_cache
import re

puzzle = Puzzle(year=2020,day=8)
print(puzzle.input_data)
```

```python
code_lines = {pc:line for pc,line in enumerate(puzzle.input_data.splitlines())}
# last_pc = len(code_lines)-1
# exec_count = defaultdict(int)

def calc_acc(program:dict) -> Tuple[int, bool]:
    exec_count = defaultdict(int)
    last_pc = len(program)-1
    pc = 0
    acc = 0
    while True:
        if exec_count[pc] >= 1:
            return acc, False
        exec_count[pc] = exec_count[pc]+1
        try:
            code, val = program[pc].split()
        except KeyError:
            return acc, True
        # if code == 'jmp' or code == 'nop':
        #     if pc + int(val) == last_pc:
        #         pc = last_pc
        if code == 'acc':
            acc = acc + int(val)
        if code == 'jmp':
            pc = pc + int(val)
            continue

        pc = pc+1

def solve_b(program:dict) -> int:
    for num, code in program.items():
        if 'jmp' in code:
            temp = program.copy()
            temp[num] = code.replace('jmp', 'nop')
            acc, terminates = calc_acc(temp)
        if 'nop' in code:
            temp = program.copy()
            temp[num] = code.replace('nop', 'jmp')
            acc, terminates = calc_acc(temp)
        
        if terminates:
            return acc



answer_a,_ = calc_acc(code_lines)
answer_b = solve_b(code_lines)
```

```python
puzzle.answer_a = answer_a
puzzle.answer_b = answer_b
```
