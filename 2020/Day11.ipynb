```python
from aocd.models import Puzzle
from typing import Tuple, Dict
from collections import defaultdict
from itertools import combinations, product, count
from functools import lru_cache
import re

puzzle = Puzzle(year=2020, day=11)
print(puzzle.input_data)
```

```python
# seats = defaultdict(lambda: '.')
seats = {}
seats.update({(x,y):val for (y, outer) in enumerate(puzzle.input_data.splitlines()) for (x,val) in enumerate(outer)})

MAX_X, MAX_Y = max(seats.keys(), key=sum)
# product([1, 0, -1], repeat=2)



def num_adjacent(seat_dict: dict, curr_seat: tuple) -> int:
    acc = 0
    offsets = list(product([1, 0, -1], repeat=2))
    offsets.remove((0, 0))
    for offset in offsets:
        if seat_dict.get((curr_seat[0]+offset[0], curr_seat[1]+offset[1]), '.') == '#':
            acc += 1
    return acc

def num_seen(seat_dict: dict, curr_seat: tuple) -> int:
    acc = 0
    offsets = list(product([1, 0, -1], repeat=2))
    offsets.remove((0, 0))
    for offset in offsets:
        for i in count(1):
            offset_pos = (curr_seat[0]+offset[0]*i, curr_seat[1]+offset[1]*i)
            if offset_pos not in seat_dict or seat_dict[offset_pos] == 'L':
                break
            if seat_dict[offset_pos] == '#':
                acc += 1
                break
    return acc

def insert_newlines(string, every=64):
    return '\n'.join(string[i:i+every] for i in range(0, len(string), every))

def solve_a(seats: dict) -> int:
    old_seats = {}
    while old_seats != seats:
        
        old_seats = seats.copy()
        for pos, val in seats.items():
            num_occupied = num_adjacent(old_seats, pos)
            if val == 'L' and num_occupied == 0:
                seats[pos] = '#'
            if val == '#' and num_occupied >= 4:
                seats[pos] = 'L'
    print(seats)
    return sum([x == '#' for x in seats.values()])

def solve_b(seats_b: dict) -> int:
    old_seats = {}
    seats = seats_b.copy()
    while old_seats != seats:
        
        old_seats = seats.copy()
        for pos, val in seats.items():
            num_occupied = num_seen(old_seats, pos)
            if val == 'L' and num_occupied == 0:
                seats[pos] = '#'
            if val == '#' and num_occupied >= 5:
                seats[pos] = 'L'
    
    return sum([x == '#' for x in seats.values()])

# answer_a = solve_a(seats)
answer_b = solve_b(seats)
```
