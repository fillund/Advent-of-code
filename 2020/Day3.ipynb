```python
from aocd.models import Puzzle
from math import prod

puzzle = Puzzle(year=2020,day=3)
print(puzzle.input_data)
```

```python
grid = {}
lines = puzzle.input_data.splitlines()
MAX_Y = len(lines)
MAX_X = len(lines[0])
for row, line in enumerate(lines):
    for col, char in enumerate(line):
        grid[(col, row)] = char


def move(pos, slope):
    return pos[0] + slope[0], pos[1] + slope[1]

# num_trees = 0
START_POS = (0,0)

# SLOPE_a = (3, 1)
SLOPES = [(3, 1), (1,1), (5,1), (7,1), (1,2)]
num_trees = {x:0 for x in range(len(SLOPES))}

for i ,slope in enumerate(SLOPES):
    current_pos = START_POS
    while current_pos[1] < MAX_Y:
        if current_pos[0] >= MAX_X:
            current_pos = current_pos[0]%MAX_X, current_pos[1]
        if grid[current_pos] == '#':
            num_trees[i] = num_trees[i] + 1
        current_pos = move(current_pos, slope)

answer_a = num_trees[0]
answer_b = prod(num_trees.values())
```

```python
puzzle.answer_a = answer_a
puzzle.answer_b = answer_b
```
