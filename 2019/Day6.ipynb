```python
from aocd.models import Puzzle
from typing import Tuple, Dict
from collections import defaultdict, deque
from functools import lru_cache
from itertools import chain
import re

puzzle = Puzzle(year=2019,day=6)
print(puzzle.input_data)
```

```python
data = puzzle.input_data.splitlines()
# data = test_input.split()
orbits = defaultdict(list)
parentdict = {'COM':None}

for line in data:
    x,y = line.split(')')
    orbits[x].append(y)
    parentdict[y] = x

@lru_cache
def count_orbits(planet: str) -> int:
    if parentdict[planet] is None:
        return 0
    else:
        return 1+count_orbits(parentdict[planet])

def bfs(start: str, goal: str) -> list:
    temp = orbits[start] + [parentdict[start]]
    temp = zip(temp, [start]*len(temp))
    q = deque(temp)
    visited = set([start])
    path = {}
    path[start] = []
    while len(q) > 0:
        v, prev = q.popleft()
        visited.add(v)
        path[v] = path[prev] + [v]
        if v == goal:
            break
        dests = [x for x in chain(orbits[v],[parentdict[v]]) if x is not None]
        for dest in dests:
            if dest not in visited and not None:
                visited.add(dest)
                q.append((dest, v))

    return path[v]


leafs = [x for x in parentdict.keys()]
answer_a = sum([count_orbits(x) for x in leafs])
answer_b = len(bfs(parentdict['SAN'], parentdict['YOU']))
```

```python
puzzle.answer_a = answer_a
puzzle.answer_b = answer_b
```
