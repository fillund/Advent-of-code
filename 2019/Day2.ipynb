```python
from aocd.models import Puzzle


puzzle = Puzzle(2019, 2)
```

```python
class IntCodeMachine:
    def __init__(self, memoryInit, ip_step):
        self.memory = memoryInit
        self.ip_step = ip_step
        self.ip = 0

    def run(self):
        while(self.memory[self.ip] != 99):
            if self.memory[self.ip] == 1:
                self.op_code1()
            elif self.memory[self.ip] == 2:
                self.op_code2()
            else:
                print('Op code not recognised')
                break
            self.ip = self.ip + self.ip_step
        print('Program done')

    def op_code1(self):
        self.memory[self.ip+3] = self.memory[self.ip+1] + self.memory[self.ip+2]

    def op_code2(self):
        self.memory[self.ip+3] = self.memory[self.ip+1] * self.memory[self.ip+2]

```

```python
data = [int(x) for x in puzzle.input_data.split(',')]
int_machine = IntCodeMachine(data, 4)
int_machine.memory[1] = 12
int_machine.memory[2] = 2
int_machine.run()
int_machine.memory
#answer_a = int_machine.memory[0]
```
